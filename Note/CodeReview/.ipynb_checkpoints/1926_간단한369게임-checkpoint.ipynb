**Input data**


100


**Output data**


1 2 - 4 5 - 7 8 - 10 11 12 - 14 15 - 17 18 - 20 21 22 - 24 25 - 27 28 - - - - -- - - -- - - -- 40 41 42 - 44 45 - 47 48 - 50 51 52 - 54 55 - 57 58 - - - - -- - - -- - - -- 70 71 72 - 74 75 - 77 78 - 80 81 82 - 84 85 - 87 88 - - - - -- - - -- - - -- 100

### 전체

1) 원소 내의 우리가 목표로 하는 3, 6, 9를 개별 원소마다 접근해서 


2) .count() 함수로 찾고,


3) count에 저장된 수 만큼을 바로 출력하는 로직

```python
T = int(input())

numbers = list(map(str,list(range(1,T+1)))) # 리스트로 1-101까지 만들고, 원소는 문자열타입


count = 0
for i in range(T): 
    count = numbers[i].count('3') + numbers[i].count('6') + numbers[i].count('9') 
    numbers[i] = int(numbers[i]) 
    if count == 0:
        print(numbers[i], end = ' ')
    elif count >= 0:
        print('-'*count, end = ' ')
```

```python
count = 0
for i in range(T):
    count = numbers[i].count('3') + numbers[i].count('6') + numbers[i].count('9')
    
    numbers[i] = int(numbers[i])
    '''
    최종 output은 문자열이 아닌 숫자이기 때문에 int로 준비하고,
    '''
    
    if count == 0: 
        '''
        count가 0라면,
        int로 준비한 원소를 그냥 출력한다.
        '''
        print(numbers[i], end = ' ')
        
    elif count >= 0:
        '''
        count가 0보다 크다면, 즉, 1이거나 2라면,
        int로 준비한 원소를 count에 담은 수 만큼 '-'으로 출력한다.
        '''
        print('-'*count, end = ' ')
```
