```python
df = pd.read_csv('code_a.csv', encoding='cp949')
df
```

```python
name = input('회사이름을 적어주세요 : ')
new_name = name.upper()
a = df[df['회사명'].isin([new_name])]
# name은 인풋 출력함수이므로 그 자체로 str, 별도의 ' 를 사용할 필요가 없다.
# print(a)
while True:
    if a.size == 0:
        print('정확한 회사명을 입력하세요')
        # 특정 문자열이 포함된 리스트 뽑기
        compony_ex = df[df.회사명.str.contains(new_name) == True]
        print(compony_ex)
        name = input('회사이름을 적어주세요 : ')
        new_name = name.upper()
        a = df[df['회사명'].isin([new_name])]
    else:
        code = a['종목코드'].values[0]
        new_code = '%06d' %code
        print(new_code)
        break
```

```python
url = 'https://finance.naver.com/item/sise_day.nhn'
result = []
for num in range(1,579):
    params = {
        'code' : new_code,
        'page' : '{}'.format(num)
    }
    resp = requests.get(url, params = params)
    soup = BeautifulSoup(resp.content, 'html.parser')
    for tr in soup.find_all('tr')[2:14]:
        new_dict = {}
        td_list = tr.find_all('td')
        if not td_list[0].text.strip():
            continue
        tdtext_list = ['날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
        for i in range(7):
            new_dict[tdtext_list[i]] = td_list[i].text.strip()
        result.append(new_dict)
from pprint import pprint
pprint(result)
```
