패턴 내의 문자들을 일일이 비교하는 방식으로 동작






비교하다가 똑같지 않은경우 
비교되는 대상은 그전 인덱스로 돌아가고
원본인덱스는 원점으로 돌아간다.


그래서앞에서 설명한 방법으로 다시 돌아간다.

```python
p = "is"
t = "This is a book~!"
M = len(p) # 2
N = len(t) # 16  #최악의 경우는 비교해야하는 경우의 수가 M*N이 된다. 

def BruteForce(p, t):
    i = 0 # t 의 인덱스
    j = 0 # p 의 인덱스
    while j < M and i < N:
        if t[i] != p[j]:
            i = i - j
            j = - 1
            
        i = i + 1
        j = j + 1
    if j == M:
        return i - M #검색 성공

    else:
        return -1 #검색 실패
    
BruteForce(p, t)
# 최악의 경우 시간복잡도는 텍스트의 모든 위치에서 패턴을 비교해야 하므로 O(MN)이된다.
```

#### 2. KMP 알고리즘

불일치가 발생한 문자열의 앞부분에 어떤 문자가 있는지 미리 알고있으므로,

```python
class KMP:
    def partial(self, pattern):
        """ Calculate partial match table: String -> [Int]"""
        ret = [0]
        
        for i in range(1, len(pattern)):
            j = ret[i - 1]
            while j > 0 and pattern[j] != pattern[i]:
                j = ret[j - 1]
            ret.append(j + 1 if pattern[j] == pattern[i] else j)
        return ret
        
    def search(self, T, P):
        """ 
        KMP search main algorithm: String -> String -> [Int] 
        Return all the matching position of pattern string P in T
        """
        partial, ret, j = self.partial(P), [], 0
        
        for i in range(len(T)):
            while j > 0 and T[i] != P[j]:
                j = partial[j - 1]
            if T[i] == P[j]: j += 1
            if j == len(P): 
                ret.append(i - (j - 1))
                j = partial[j - 1]
            
        return ret



def test():
    p1 = "aa"
    t1 = "aaaaaaaa"

    kmp = KMP()
    assert(kmp.search(t1, p1) == [0, 1, 2, 3, 4, 5, 6])
    #https://wikidocs.net/21050

    p2 = "abc"
    t2 = "abdabeabfabc"

    assert(kmp.search(t2, p2) == [9])

    p3 = "aab"
    t3 = "aaabaacbaab"

    assert(kmp.search(t3, p3) == [1, 8])

    print("all test pass")
```

1. 고지식한 검색(순차 검색, sequential search)
    - 자료들의 리스트에서 키 값을 찾기위해 첫 번째 자료부터 비교하면서 진행

```python
def sequentialsearch(a, n, key): # a -> list 
    i = 0
    while i < n and a[i] != key: # n이 1보다 크면 트루, a 리스트와 키가 다르면 트루
        i = i + 1 # 그러면 i에 1을 더한다.
        
    if i < n: 
        return i # 성공하면 키값의 인덱스 값을 반환한다.
    else:
        return -1 # 실패
```

```python
# 두 원소의 교환을 통해 생성하는 순열
# 재귀 호출을 통한 순열 생성

# a[] : list stored with data
# n : number of arguments
# k : selected number of arguments


def perm(n, k):
    if k == n:
        return a
    else:
        for i in range(k, n):
            a[k], a[i] = a[i], a[k]
            perm(n, k+1)
            a[k], a[i] = a[i], a[k]
```
